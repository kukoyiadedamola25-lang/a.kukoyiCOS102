```python
import tkinter as tk
from tkinter import messagebox

class Delivery:
    def __init__(self, location, weight):
        self.location = location
        self.weight = weight

    def calculate_fee(self):
        """Calculates fee based on location and weight rules."""
        if self.location == "PAU":
            if self.weight >= 10:
                return 2000
            else:
                return 1500
        
        elif self.location == "Epe":
            if self.weight >= 10:
                return 5000
            else:
                return 4000
        
        return 0

class DeliveryApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Logistics Price Calculator")
        self.root.geometry("350x300")
        self.root.config(padx=20, pady=20)

        # Location Selection
        tk.Label(root, text="Select Delivery Location:", font=("Arial", 10, "bold")).pack(pady=5)
        self.location_var = tk.StringVar(root)
        self.location_var.set("PAU")  # Default value
        self.location_menu = tk.OptionMenu(root, self.location_var, "PAU", "Epe")
        self.location_menu.pack(pady=5)

        # Weight Input
        tk.Label(root, text="Enter Package Weight (kg):", font=("Arial", 10, "bold")).pack(pady=5)
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack(pady=5)

        # Calculate Button
        self.calc_button = tk.Button(
            root, 
            text="Calculate Price", 
            command=self.show_price,
            bg="#2ecc71", 
            fg="white",
            font=("Arial", 10, "bold")
        )
        self.calc_button.pack(pady=20)

    def show_price(self):
        try:
            # Get data from UI
            location = self.location_var.get()
            weight = float(self.weight_entry.get())

            if weight < 0:
                raise ValueError("Weight cannot be negative.")

            # Create an instance of the Delivery class (OOP)
            delivery_session = Delivery(location, weight)
            price = delivery_session.calculate_fee()

            # Display result
            messagebox.showinfo("Quote", f"Total Delivery Charge: N{price}")
            
        except ValueError:
            messagebox.showerror("Input Error", "Please enter a valid numerical weight.")

# Main loop
if __name__ == "__main__":
    root = tk.Tk()
    app = DeliveryApp(root)
    root.mainloop()
```
