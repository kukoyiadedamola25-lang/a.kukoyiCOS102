```python
import tkinter as tk
from tkinter import messagebox
import random

class Employee:
    def __init__(self):
        # The database of authorized personnel
        self.employee_list = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", 
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James", 
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso", 
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]
        self.tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]
        self.attendance_log = []

    def check_employee(self, name):
        """Verify if the name exists in the database."""
        if name in self.employee_list:
            return True
        return False

    def take_attendance(self, name):
        """Log the employee's name for the session."""
        self.attendance_log.append(name)
        print(f"Attendance recorded for: {name}")

    def assign_task(self):
        """Randomly select a task from the list."""
        # Using random.randint as requested to pick an index
        index = random.randint(0, len(self.tasks) - 1)
        return self.tasks[index]

    def refuse_access(self):
        """Return a polite refusal message."""
        return "Access Denied. You are not registered in the employee database."

class EmployeeApp:
    def __init__(self, root):
        self.system = Employee()
        self.root = root
        self.root.title("Jane's Delivery Logistics System")
        self.root.geometry("400x250")

        # UI Elements
        self.label = tk.Label(root, text="Please Enter Your Full Name:", font=("Arial", 10))
        self.label.pack(pady=20)

        self.name_entry = tk.Entry(root, width=30)
        self.name_entry.pack(pady=5)

        self.submit_btn = tk.Button(root, text="Login / Check Task", command=self.process_login, bg="#4CAF50", fg="white")
        self.submit_btn.pack(pady=20)

    def process_login(self):
        entered_name = self.name_entry.get().strip()

        if self.system.check_employee(entered_name):
            # Process Employee
            self.system.take_attendance(entered_name)
            task = self.system.assign_task()
            
            success_msg = f"Welcome, {entered_name}!\n\nAttendance: Recorded\nToday's Task: {task}"
            messagebox.showinfo("Access Granted", success_msg)
        else:
            # Refuse Access
            failure_msg = self.system.refuse_access()
            messagebox.showerror("Access Denied", failure_msg)

# Main execution
if __name__ == "__main__":
    root = tk.Tk()
    app = EmployeeApp(root)
    root.mainloop()
```
