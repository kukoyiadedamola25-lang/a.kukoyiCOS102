```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox

# 1. Load the data
try:
    df = pd.read_csv('GIG-logistics.xlsx - in.csv')
except FileNotFoundError:
    print("Error: File not found.")

def check_employee():
    # Get the Name and Department from the input boxes
    typed_name = name_entry.get().strip()
    typed_dept = dept_entry.get().strip()
    
    # Filter: Look for a row where Name AND Department match
    # we use .str.lower() to make the search not care about Capital Letters
    match = df[(df['Name'].str.lower() == typed_name.lower()) & 
               (df['Department'].str.lower() == typed_dept.lower())]

    if not match.empty:
        # If a match is found, they are an employee!
        # Grab their specific info from that row
        employee_info = match.iloc[0] # Get the first (and likely only) match
        
        msg = f"Welcome, {employee_info['Name']}!\n"
        msg += f"Role: {employee_info['Role']}\n"
        msg += f"Location: {employee_info['Location']}"
        
        messagebox.showinfo("Success", msg)
    else:
        # If no match is found
        messagebox.showerror("Access Denied", "Employee not found in this department.")

# --- GUI Setup ---
root = tk.Tk()
root.title("GIG Logistics - Employee Verification")
root.geometry("400x300")

tk.Label(root, text="Employee Verification System", font=("Arial", 12, "bold")).pack(pady=10)

# Name Input
tk.Label(root, text="Enter Full Name:").pack()
name_entry = tk.Entry(root, width=30)
name_entry.pack(pady=5)

# Department Input
tk.Label(root, text="Enter Department:").pack()
dept_entry = tk.Entry(root, width=30)
dept_entry.pack(pady=5)

# Button
check_btn = tk.Button(root, text="Verify Employee", command=check_employee, bg="green", fg="white")
check_btn.pack(pady=20)

root.mainloop()
```
