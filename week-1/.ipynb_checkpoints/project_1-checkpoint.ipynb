```python
# Simple Interest Program

P = float(input("Enter the Principal amount: "))
R = float(input("Enter the interest rate (%): "))
T = float(input("Entrer the time (years): "))

A = P * (1 + (R/100) * T)

print("The total amount is:", A)

# Compound Interest Program

P = float(input("Enter the principal amount: "))
R = float(input("Enter the interest rate (%): "))
n = int(input("Enter number of times interest is compounded per year: "))
t = float(input("Enter time in years: "))

A = P * (1 + (R/(100*n)))**(n*t)

print("The compound amount is:", A)

# Annuity Plan Program

PMT = float(input("Enter payment amount: "))
R = float(input("Enter interest rate (%): "))
n = int(input("Enter number of payments per year: "))
t = float(input("Enter number of years: "))

A = PMT * (((1 + (R/(100*n)))**(n*t) - 1) / (R/(100*n)))

print("Future value of annuity:", A)
```
